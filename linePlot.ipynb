```python
import matplotlib.pyplot as plt

x = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']
y = [300, 450, 330, 500, 400]
y1 = [400, 540, 480, 430, 550]

plt.plot(x, y, marker='*',ls = '--', color='r', label = "Week1")
plt.plot(x, y1, marker='o',ls = '-', color='g', label = "Week2", alpha = 0.5)
plt.legend()
plt.show()
```
