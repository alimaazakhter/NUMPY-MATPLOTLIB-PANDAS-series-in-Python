```python
import matplotlib.pyplot as plt

x = ["part 1",' part 2', 'part 3', 'part 4', 'part 5']
y = [89,97,77,80, 83]
color = ['darkred', 'green', 'blue', 'yellow', 'purple']
plt.bar(x, y, color=color, edgecolor='black')
plt.xlabel('Parts of the Harry Potter series', fontsize=17, color='orange')
plt.ylabel('Popularity', fontsize=17, color='orange')
plt.title('Popularity of Harry Potter series', fontsize=20)
plt.show()
```
