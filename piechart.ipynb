```python
import matplotlib.pyplot as plt

brands = ['Nokia', 'Samsung', 'Apple', 'Oneplus', 'Redmi']
x = [12, 46, 78, 33, 22]
ex = (0, 0 , 0, 0, 0.1)
plt.pie(x, labels=brands, explode=ex, shadow=True, autopct='%.2f%%')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

data = pd.read_excel("Expense.xlsx")
df = pd.DataFrame(data)
groupedby = df.groupby('Payment Mode')['Amount'].sum()
print(groupedby)
plt.pie(groupedby.values, labels=groupedby.index, autopct='%.2f%%')
plt.show()
print(df)
```
